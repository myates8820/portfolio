```python
raw_save = 'nba_2019.csv'

if not os.path.exists(raw_save):
    raw_games = pd.read_html('https://www.basketball-reference.com/leagues/NBA_2019_games.html')
    raw_table = raw_games[0]
else:
    raw_table = pd.read_csv(raw_save)
```

```python
#turns out the tables are only for the month of October. Going to write a loop to build the list for the complete season
raw_table.head()
```

```python
# creating a loop to create links for the read_html function
months = ['october', 'november', 'december', 'january', 'february', 'march', 'april', 'may', 'june']
links = []
for x in months:
    link_month = 'https://www.basketball-reference.com/leagues/NBA_2019_games-' + x + '.html'
    links.append(link_month)
links
    
```

```python
if not os.path.exists(raw_save):
    for x in links[1:]:
        month_games = pd.read_html(x)
        month_table = month_games[0]
        raw_table = pd.concat([raw_table,month_table], ignore_index = True)
        raw_table.to_csv('nba_2019.csv')
else:
    raw_games = pd.read_csv(raw_save)
```

```python
# Function to which columns are missing data and how much
print("The following columns are missing data.")
print()
missing_data = []
for x in list(raw_table.columns):
    amt_missing = raw_table[x].isnull().sum()
    if amt_missing > 0:
        missing_data.append(x)
    pct_missing = round((amt_missing/raw_table.shape[0])*100,1)
    if pct_missing>0:
        print(f'{x} is missing {pct_missing}% of its data.')
if len(missing_data)==0:
    print('There are no null values in the dataset.')
```

```python
# Going to fix column names and convert dates and times
cols = {'Unnamed: 0': 'id', 'Date': 'date', 'Start (ET)': 'start', 'Visitor/Neutral': 'away_team', 'PTS': 'points_away',
        'Home/Neutral': 'home_team',
       'PTS.1': 'points_home', 'Unnamed: 7': 'ot', 'Attend.': 'attendance'}

raw_table = raw_table.rename(columns = cols)
```

```python
raw_table['date'] = raw_table['date'].apply(lambda x: pd.to_datetime(x))
```

```python
# turning start time into 24:00 in order to simplify sorting
def twentyfourclock(x):
    if len(x) == 6:
        new_base = int(x[0:2])+12
        return str(new_base)+':00'
    else:
        new_base = int(x[0])+12
        return str(new_base)+':00'
raw_table['start'] = raw_table['start'].apply(twentyfourclock)
```

```python
raw_table = raw_table.astype({'points_away': 'int64', 'points_home': 'int64', 'attendance': 'int64'})
```

```python
# Let's ask a final question, are there such things as hot streaks? 
```

```python
regular_teams['team_wins'] = (regular_teams.sort_values(['team','date'])
                                                           .groupby('team')['win'].cumsum(axis=0))
```

```python
regular_teams['team_games'] = (regular_teams.assign(one = lambda x: 1)
                               .sort_values(['team','date'])
                               .groupby('team')['one'].cumsum(axis=0))
```

```python
regular_teams.sort_values(['team','date']).head()
```

```python
regular_teams = (regular_teams.assign(running_pct = lambda x: x.team_wins / x.team_games))
```

```python
regular_teams.sort_values(['team','date']).head()
```

```python
# now need to create a column to check if teams won their previous game
df = regular_teams.sort_values(['team', 'date']).reset_index()
df[['team_lagged','win_lagged']] = df[['team', 'win']].shift(periods=1, fill_value = False)
df['win_count'] = df['win_lagged'].cumsum(axis=0)
```

```python
# using a mask to reset values in win_streak column
df['win_count'] = df['win_lagged'].cumsum(axis=0)
df['win_streak'] = (df['win_count'].sub(df['win_count'].mask(df['win_lagged']).
                                        ffill().fillna(0)).astype(int))
```

```python
sns.lineplot(x = df.groupby('win_streak')['win'].mean().reset_index()['win_streak'], 
             y = df.groupby('win_streak')['win'].mean().reset_index()['win'])
```

```python
# To see how meaningful our data is, let's see how often these streaks occur

sns.countplot(x=df[df['win_streak']>0]['win_streak'])
```

```python
# Well there we go. We can see that teams do have a higher tendency to win their next game if they're on a win streak of 3-5
# games, whereas the percantages fluctuate afterwards (mostly due to lack of data). Nonetheless, this is an interesting
# insight
```

```python
# Let's ask another question. Does average attendance size go up as a win streak continues for the home team? For the away team?
# How does attendance interact with team performance?
```

```python
attendance_wins = df.merge(attendance[['id', 'attendance']], how = 'inner', on='id', 
                                      left_index=True, right_index=True)
attendance_wins = attendance_wins.drop(['index', 'win_home', 'team_lagged', 'win_lagged', 'win_count'], axis=1)
```

```python
sns.lineplot(x=attendance_wins.groupby(['win_streak', 'home_away'])['attendance'].mean().reset_index()['win_streak'],
            y=attendance_wins.groupby(['win_streak', 'home_away'])['attendance'].mean().reset_index()['attendance'],
           hue = attendance_wins.groupby(['win_streak', 'home_away'])['attendance'].mean().reset_index()['home_away'])
```

```python
# Let's take a look at home attendance by number of games played by the home team

sns.lineplot(x = attendance_wins['team_games'],
             y = attendance_wins[attendance_wins['home_away']=='home_team'].groupby('team_games')['attendance'].mean())
```

```python
attendance_wins['rank'] = (attendance_wins.groupby('team')
            .tail(1)
            .sort_values('running_pct'))['team_wins'].rank(ascending=False)
rankings = (attendance_wins.groupby('team')
            .tail(1)
            .sort_values('running_pct')[['team','rank',]]
            .reset_index()
            .drop('index', axis=1))
rankings = rankings.merge(attendance_wins.groupby('team')['attendance'].mean().reset_index(), how = 'inner', on='team')
```

```python
plt.subplots(figsize = (12,5))
sns.barplot(x=rankings['rank'],
           y=rankings['attendance'])
```
