```python
# Looks like both have a pattern of two main sections. For lotshape, IR and regular will be dummied
# For landcontour, BnK and the rest will be dummied. This is due to low std of bnk with large difference, while others are
# small difference with larger std.
# Created function to accept as arguments the desired dummy variable, the column from which to make it, and the dataframe which
# to make it in.
# Potential modifications for later on are for two dataframes, perhaps once I have statistical mean tests, make an automatic
# program to decide if I should or should not create a dummy variable or one hot encode.


def twotype_dummy(desired_dummy,col,df):
    if col not in list(df.columns):
        print(col, " is not a column in entered dataframe.")
        print("Please correct the inputs and try again.")
        
    elif  desired_dummy not in list(df[col].unique()):
        print(desired_dummy, " is not a type for the column: ", col)
        print("Please correct the inputs and try again.")
    
    else:
        new_df_name = col + desired_dummy[0:3]
        def type_singledummy(x):
            if x == desired_dummy:
                return 1
            else:
                return 0
        df[new_df_name] = df[col].apply(type_singledummy)
        
twotype_dummy('Reg','LotShape',houses1)
houses1.drop('LotShape',axis=1,inplace=True)
```
